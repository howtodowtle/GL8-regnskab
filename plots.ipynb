# Plots

Import matplotlib and Gorgi's methods:

```python
import gorgireader
import gorgimessages
import matplotlib.pyplot as plt
%matplotlib inline
```

Read in the file and add a variable that determines in which "saldo class" each person falls:

```python
regnskab = gorgireader.getWorkbook()
gorgimessages.whichMessage(regnskab)
```

## Individual saldos

Create a horizontal bar plot for the individual saldos:

```python
fig, ax = plt.subplots(figsize = (12, 8))

ax.barh(range(1, 24), regnskab["ny_saldo"], align = 'center', color = regnskab.msg_type.map({-1: "r", 0: "0.75", 1: "g"}), ecolor = 'black')
#plt.axis([-2050, 2050, -1, 25])
ax.set_yticks(range(1, 24))
ax.set_yticklabels(regnskab["name"])
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Din saldo')
plt.vlines(0, ymin = 0, ymax = 24, colors = 'k', linestyles = '-')
plt.vlines(300, ymin = 0, ymax = 24, colors = 'k', linestyles = ':')
ax.set_title('Venligst overfør op til +300 kr. (eller mere)')

for i, v in enumerate(regnskab["ny_saldo"]):
    ax.text(v + 20, i + 1.2, str(round(v)), color="k")

plt.show()
```
